# Checking and Handling Missing data

```python

# Percentage of missing values .
missing_values_percentage = startup_df.isnull().mean()*100

missing_values_percentage
```

```python
# Visualise missing data .
msno.bar(startup_df);
```

It seems we have a lot of missing data in the closed_at column,  63.705309 % of data is missing.   53.412784 % of data is missing in unmaned_6 column,age_first_milestome has 16.468039 % of missing data ,whereas age_last_milestone_year has  16.468039 % of the data missing .

How do we solve this problem ?.Null values at closed_at column might mean that the  company is still operating . We will create a new column called last_date and drop this one . We will also remove all the unnamed columns , there is nothing we can derive from these columns since we do not know what the records of that column entail .

However, it appears to me that column age_first_funding and age_first_milestone_year have little data missing , thus we can safely work with it at this point . 

```python
# Handle missing values .

# Drop columns with too much missing data .
startup_df.drop('closed_at',axis=1,inplace=True)

# Drop unammed columns since we they wont be of any value
startup_df.drop(['Unnamed: 0','Unnamed: 6' ],axis=1,inplace=True)


# Confirm removed columns
startup_df.head()
```

```python
# Check for duplications
startup_df.duplicated().sum()
```

```python
# Check if state_code column is the same with column called state_code1

# # Find equality within states code .
similar = startup_df['state_code'].equals(startup_df['state_code.1'])
print(f'Are there equal ? : {similar}')


# Loop through all the state and find similarity .
def matching_column(column,colum_to_match):
    not_similar = []
    for i,j in zip(startup_df[column],startup_df[colum_to_match]):
        if i == j :
            continue
        else :
            not_similar.append((i,j))
    print(f'Number of disimilar values : {len(not_similar)}\nThe disimilar value(s) are :{not_similar}') 
    
matching_column('state_code','state_code.1')
```

Using the duplicate function has proven to us that there are not duplicates in our dataset . If you  look closely to our columns , there seem to be some similarities between state_code.1 and state_code .After using the equals function , it will seem at first glance that there are indeed disimilar columns . 

However through the function that we have created we  notice that there is only 1 entrie(NAN)which is not similar to other entries . Upon this finding , we can conclude that the two columns are similar and we can safely drop the other column . We will also apply this method to other column which we suspect to be similar .

```python
# Checking for matching column
matching_column('object_id', 'id')
```

```python
# Drop similar column and useless columns

startup_df.drop(['state_code.1','object_id','id','latitude', 'longitude', 'zip_code'],axis=1,inplace=True)

```

```python
# Reset options to view all the columns in the dataset .
pd.set_option('display.max.columns',None)
startup_df.head()
```

```python
# Check for date inconsitencies and data types 
print(startup_df['founded_at'].dtype)
print(startup_df['first_funding_at'].dtype)
print(startup_df['last_funding_at'].dtype)
```

Columns with dates are of datatype object . We do not want this ,thus we are going to transform these columns to datetime . This will make our analysis easier later on  .

```python
# Transforn columns with date to their correct datatype .

# Convert dates from objects to datetime
startup_df['founded_at'] = pd.to_datetime(startup_df['founded_at'])
startup_df['first_funding_at'] = pd.to_datetime(startup_df['first_funding_at'])
startup_df['last_funding_at'] = pd.to_datetime(startup_df['last_funding_at'])

#Check if columns were converted.
startup_df[['last_funding_at','last_funding_at','founded_at']].info()
```

# Correlation

```python
plt.figure(figsize=(20,10)) # Plot and show the correlation of feautures .
sns.heatmap(startup_df.corr(),annot=True, cmap='viridis');
```

```python
# Anaylise correlation.

# Plot the first graph 
plt.figure(figsize=(12,7))
plt.subplot(2,2,1)
sns.regplot(x=startup_df["age_first_funding_year"], y=startup_df["age_last_funding_year"])

# Plot the second graph 
plt.subplot(2,2,2)
sns.regplot(x=startup_df["age_first_milestone_year"], y=startup_df["age_last_milestone_year"])

# Plot the third graph 
plt.subplot(2,2,3)
sns.regplot(x=startup_df["milestones"], y=startup_df["funding_total_usd"])

# Plot the fourth graph 
plt.subplot(2,2,4)
sns.regplot(x=startup_df["milestones"], y=startup_df["funding_rounds"])

plt.show()
```

Insights :


There is a strong realtionship between age last funding year and the age first funding year . This is the same with age last milestones and age last milestone year.

Startups with 2 and 3 milestones will get more funding rounds (up to 8) . On the other hand there seem to be a weak relationship between the milestones which the company has achieved with the total amount of money it raised .
